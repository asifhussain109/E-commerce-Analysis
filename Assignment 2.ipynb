```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
color = sns.color_palette()
```

```python
df = pd.read_csv('D:\work\python\Python Assignment 2_BI\Ecommerce - UK Retailer.csv',encoding= 'unicode_escape')
df.head(10)
```

# Data Cleanning

```python
df.drop_duplicates()
```

```python
df_new = df[(df['Quantity'] >= 0)]
```

```python
df_new.info()
```

# Update the data

```python
df_new['InvoiceDate'] = df_new['InvoiceDate'].astype('datetime64[ns]')
df_new['Year'] = df_new.InvoiceDate.dt.year
df_new['Month'] = df_new.InvoiceDate.dt.month
df_new['Day']=(df_new.InvoiceDate.dt.dayofweek)+1 
df_new['Hour'] = df_new.InvoiceDate.dt.hour
df_new['Revenue'] = df_new['Quantity'] * df_new['UnitPrice']
df_new.head()
```

# 10. How many orders per month?

```python
y = df_new.groupby('InvoiceNo')['Month'].unique().value_counts().sort_index().plot(kind='bar',color=color[0],figsize=(14,8))
y.set_xlabel('Month',fontsize=14)
y.set_ylabel('Number of Orders',fontsize=14)
y.set_title('Number of orders for different Months (1st Dec 2010 - 9th Dec 2011)',fontsize=18)
y.set_xticklabels(('Dec_10','Jan_11','Feb_11','Mar_11','Apr_11','May_11','Jun_11','July_11','Aug_11','Sep_11','Oct_11','Nov_11','Dec_11'), rotation='horizontal', fontsize=11)
plt.show()
```

# 11. How many orders per day?

```python
ir = df_new.groupby('InvoiceNo')['Day'].unique().value_counts().sort_index().plot(kind='bar',color=color[0],figsize=(14,5))
ir.set_xlabel('Day',fontsize=18)
ir.set_ylabel('Orders',fontsize=18)
ir.set_xticklabels(('Mon','Tue','Wed','Thur','Fri','Sun'), rotation='horizontal', fontsize=15)
plt.show()
```

# 12. How many orders per hour?

```python
ir = df_new.groupby('InvoiceNo')['Hour'].unique().value_counts().iloc[:-1].plot(kind='bar',color=color[0],figsize=(15,6))
ir.set_xlabel('Hour',fontsize=13)
ir.set_ylabel('Number of Orders',fontsize=13)
ir.set_title('Number of Hours',fontsize=13)
ir.set_xticklabels(range(6,21), rotation='horizontal', fontsize=13)
plt.show()
```

# 13. How many orders for each country?

```python
group_country = df_new.groupby('Country')['InvoiceNo'].count().sort_values()
del group_country['United Kingdom']

# number of unique customers in each country (without UK)
plt.subplots(figsize=(15,8))
group_country.plot(kind='barh', fontsize=14, color=color[0])
plt.xlabel('Number of Orders', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.title('Number of Orders for different Countries', fontsize=14)
plt.show()
```

# 14. Orders trend across months

```python
fig = plt.figure(figsize=(15, 7))
x1 = fig.add_subplot(111)

x1.set_xlabel('Month',fontsize=15)
x1.set_ylabel('Total Amount Spent ($)',fontsize=15)
x1.set_title('Total Amount Spent for Months',fontsize=15)

x1.bar(df_new.groupby(by=['Month'])['Revenue'].sum().index.tolist()[1:],
        df_new.groupby(by=['Month'])['Revenue'].sum()[1:],
        alpha=0.85,#If you want to make the graph plot more transparent, then you can make alpha less than 1, such as 0.5 or 0.25.
        label='Amount Spent by Month')

x2 = x1.twinx()# tread line

x2.set_ylabel('Percentage Change of Total Amount Spent (%)',fontsize=15)

#pct_change()Percentage change between the current and a prior element.
x2.plot(df_new.groupby(by=['Month'])['Revenue'].sum().index.tolist()[1:], 
         df_new.groupby(by=['Month'])['Revenue'].sum()[1:].pct_change().fillna(0)*100,
         label='Percentage change of total amount spent (%)',
         color='red')

x1.legend(loc='upper left')
x2.legend(loc='upper right')

fig.tight_layout()
plt.show()
```
